# Get Kit a Red Wagon!

I want to get kit a Radio Flyer wagon. New ones are expensive. I need to regularly check Craigslist, but that's boring. This Python script will search Craigslist for Radio Flyers for sale near me and email me the results. I scheduled it with Windows Task Scheduler and now get a list of Radio Flyers for sale in my area in my email each morning.

```python
from urllib import urlopen
from bs4 import BeautifulSoup as bs4
from collections import defaultdict

import pprint
import requests

import gmail
from gmail import Message
```

```python
url_base = 'https://pueblo.craigslist.org/search/baa'  # kids and babies
params = dict(query=['Radio Flyer'], search_distance=50, postal=81007)
response = requests.get(url_base, params=params)
html = bs4(response.text, 'html.parser')
results = html.find_all(class_='result-info')

print 'url:'
print response.url
print
print 'results: {}'.format(len(results))
```

```python
def findWagon(results):
    wagonDict = defaultdict(dict)
    for i in range(len(results)):
        try:
            wagonDict[i] = {
            'date': results[i].find(class_='result-date')['datetime'],
            'cost': results[i].find(class_='result-price').text,
            'link': results[i].find(class_='result-title hdrlnk')['href'],
            'type': results[i].find(class_='result-title hdrlnk').text,
            'city': results[i].find(class_='nearby')['title'],
            'area': results[i].find(class_='nearby').text,
            'pics': ('pic' in results[i].text)
            }
        except:
            pass
    return wagonDict
```

```python
email = gmail.GMail(my_email, my_password)
email.connect()
```

```python
wagons = findWagon(results)
if wagons:
    message = ''
    for i, wagon in wagons.items():
        entry = \
        'Title: {}\nDate: {}\nPrice: {}\nCity: {}\nArea: {}\nPictures: {}\nLink: {}'\
        .format(wagon['type'], wagon['date'], wagon['cost'], wagon['city'],
        wagon['area'], wagon['pics'], wagon['link'])
        print entry
        print
        message += entry + '\n\n'
    
    payload = Message(subject='Wagons', to=my_email, text=message)
    email.send(payload)
```
